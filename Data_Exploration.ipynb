## Data Exploration

```python
# EDA for ELLIPSE Corpus (train.csv)

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv("data/train.csv")

# Display basic information
print("Dataset Shape:", df.shape)
print("nColumn Types:", df.dtypes)
```

```python
# Show missing values
print("\nMissing Values:\n", df.isnull().sum())
```

```python
# Show basic statistics
print("\nDescriptive Statistics:\n", df.describe())
```

```python
import numpy as np
```

```python
# Distribution of each scoring feature
score_columns = ['cohesion', 'syntax', 'vocabulary', 'phraseology', 'grammar', 'conventions']

for col in score_columns:
    plt.figure(figsize=(6, 4))
    value_counts = df[col].value_counts().sort_index()
    plt.bar(value_counts.index, value_counts.values, width=0.4)
    plt.title(f'Bar Chart of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.grid(axis='y')
    plt.xticks(value_counts.index)
    plt.show()
```

```python
# Correlation heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(df[score_columns].corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Between Score Features")
plt.show()
```

```python
# Pairplot to explore relationships
sns.pairplot(df[score_columns])
plt.suptitle("Pairplot of Score Features", y=1.02)
plt.show()
```
