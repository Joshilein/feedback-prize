```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("data/feature_engineered/train_vars.csv", engine="python", on_bad_lines="skip")
print(df)
```

```python
columns_to_drop = ['text_id', 'full_text', 'cohesion', 'syntax', 
                   'vocabulary', 'phraseology', 'grammar', 'conventions']

features = df.drop(columns=columns_to_drop)
```

```python
correlation_matrix = features.corr(method='pearson')  
```

```python
plt.figure(figsize=(18, 15))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap="coolwarm")
plt.title("Feature-Correlations (pearson)")
plt.show()
```

```python
correlation_matrix = features.corr(method='spearman')
```

```python
plt.figure(figsize=(18, 15))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap="coolwarm")
plt.title("Feature-Correlations (spearman)")
plt.show()
```
